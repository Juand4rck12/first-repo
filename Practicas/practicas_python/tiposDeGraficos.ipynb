```python
# Librerías clave
import matplotlib.pyplot as plt  # Gráficos básicos
import seaborn as sns            # Visualización estadística
import pandas as pd              # Manejo de datos
```

```python
datos = pd.read_csv('DatosSeguros.csv')
datosLimpios = datos.dropna().drop_duplicates()
datosLimpios.isnull().sum()
```

```python
# Código histograma
plt.figure(figsize=(10,6))
sns.histplot(data=datosLimpios, x='edad', bins=20, kde=True, color='skyblue')
plt.title('Distribución de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Código
plt.figure(figsize=(10,6))
sns.boxplot(data=datosLimpios, x='region', y='valor_seguro')
plt.title('Distribución del Valor del Seguro por Región')
plt.xlabel('Región')
plt.ylabel('Valor del Seguro ($)')
plt.show()
```

```python
# Código
plt.figure(figsize=(10,6))
sns.scatterplot(data=datosLimpios, x='edad', y='valor_seguro', hue='fumador', 
               palette=['green','red'], size='imc', sizes=(20,200))
plt.title('Relación entre Edad y Valor del Seguro')
plt.xlabel('Edad (años)')
plt.ylabel('Valor del Seguro ($)')
plt.show()
```

```python
# Código
plt.figure(figsize=(10,6))
sns.barplot(data=datosLimpios, x='region', y='valor_seguro', hue='sexo')
plt.title('Valor Promedio del Seguro por Región y Sexo')
plt.xlabel('Región')
plt.ylabel('Valor Promedio ($)')
plt.legend(title='Sexo')
plt.show()
```

```python
# Código
corr = datosLimpios.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de Correlación')
plt.show()
```

```python
# Violin Plot: Comparación del valor del seguro por región y sexo
plt.figure(figsize=(12, 6))
sns.violinplot(data=datosLimpios, x='region', y='valor_seguro', hue='sexo',
               palette={'F': 'pink', 'M': 'lightblue'}, split=True)
plt.title('Distribución del Valor del Seguro por Región y Sexo', fontsize=14)
plt.xlabel('Región', fontsize=12)
plt.ylabel('Valor del Seguro ($)', fontsize=12)
plt.legend(title='Sexo', loc='upper right')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pair Plot: Relaciones entre variables numéricas
variables = ['edad', 'imc', 'hijos', 'valor_seguro']
sns.pairplot(data=datosLimpios[variables + ['fumador']], 
             hue='fumador', 
             palette={'yes': 'red', 'no': 'green'},
             corner=True,  # Mostrar solo la mitad inferior
             diag_kind='kde',  # Gráficos diagonales como densidad
             plot_kws={'alpha': 0.6, 's': 30})  # Personalización
plt.suptitle('Relaciones entre Variables Clave', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=datosLimpios, x='fumador')
plt.title("Proporcion de fumadores")
plt.xlabel("¿Es fumador?")
plt.ylabel("Cantidad")
plt.show()
```
